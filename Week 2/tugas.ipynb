```python
import pandas as pd

data = 'wbc.csv' # path dataset
df = pd.read_csv(data) # load dataset
```

# Memisahkan variable
Memisahkan dengan variabel yang tidak relevan untuk digunakan yaitu id dan Unnamed:32 (ini dikarenakan berisi NaN)

```python
df_cleaned = df.drop(columns=['id', 'Unnamed: 32'])
df_cleaned.head()
     
```

```python
df_cleaned.isnull().sum()
```

# Melakukan encoding

```python
from sklearn.preprocessing import OneHotEncoder

#Inisialisasi OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

#Transformasi kolom "diagnosis" menggunakan OneHotEncoder
diagnosis_encoded = encoder.fit_transform(df_cleaned[['diagnosis']])

#Membuat DataFrame baru hasil encode
diagnosis_encoded_df = pd.DataFrame(diagnosis_encoded, columns=encoder.get_feature_names_out(['diagnosis']))

#Menggabungkan DataFrame yang hasil encode dengan DataFrame asli
df_encoded = pd.concat([df_cleaned, diagnosis_encoded_df], axis=1)

#Hapus kolom asli "diagnosis" karena kita telah menggantinya dengan versi yang di-encode
df_encoded = df_encoded.drop(columns=['diagnosis'])

df_encoded.head()
```

# Standarisasi

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
df_encoded.iloc[:,0:-2] = std.fit_transform(df_encoded.iloc[:,0:-2]) # mengambil kolom diagnosis yang berada di 2 kolom terakhir
df_encoded.head()
```

# Stratified Split

```python
from sklearn.model_selection import train_test_split

# X adalah fitur-fitur atau variabel, y adalah label
x = df_encoded.drop(columns=['diagnosis_B', 'diagnosis_M'])
y = df_encoded[['diagnosis_B', 'diagnosis_M']]

# Membagi data dengan perbandingan 80:20 (train dan test)
x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.2, random_state=42)

# Split lagi antara validasi dan testing dari x_temp dan y_temp
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.5, random_state=0)

# Mencetak jumlah data latih, data validasi, dan data uji
print("Total data training:", len(x_train))
print("Total data validation:", len(x_val))
print("Total data testing:", len(x_test))
```
