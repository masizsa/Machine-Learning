```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import statsmodels.api as sm
```

# Import Data

```python
df = pd.read_csv('insurance.csv')
df.head()
```

# Encoding

```python
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df['smoker'] = le.fit_transform(df['smoker'])
df['region'] = le.fit_transform(df['region'])

df.head()
```

# Pilih Data

```python
x = df[['age', 'bmi', 'smoker']]
y = df['charges']
```

# Split Data

```python
# Pembagian data latih dan data uji dengan proporsi 7:3
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scalers = ['age', 'bmi', 'smoker']
x_train_scaled = x_train.copy()
x_test_scaled = x_test.copy()

for col in scalers:
    scaler = StandardScaler()
    x_train_scaled[col] = scaler.fit_transform(x_train[[col]])
    x_test_scaled[col] = scaler.transform(x_test[[col]])
```

```python
print("Original X_train:")
print(x_train.head())

print("\nScaled X_train:")
print(x_train_scaled.head())
```

# Membuat multiple linear regresion model

```python
linear_model = LinearRegression()
linear_model.fit(x_train_scaled, y_train)
```

# Pelatihan Model

```python
y_pred = linear_model.predict(x_test_scaled)

import statsmodels.api as sm

X_train_sm = sm.add_constant(x_train_scaled)
lr = sm.OLS(y_train, X_train_sm).fit()
lr.summary()
```

```python
lr.params
```

# Evaluasi

```python

mse = mean_squared_error(y_test, y_pred)
r2_score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f'MSE Score : {mse}')
print(f'R2 Score : {r2_score}')
print(f'MAE Score : {mae}')
```
