```python
# Import Library
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
dt = pd.read_csv('dataset/mushrooms.csv')
dt.head()
```

```python
dt.columns
```

```python
dt.isnull().sum()
```

```python
X = dt.drop('class', axis=1)
y = dt['class']

X = pd.get_dummies(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
dt = DecisionTreeClassifier(criterion='entropy', random_state=42)

rf = RandomForestClassifier(random_state=42)

param_grid_dt = {
    'max_depth' : [None, 10, 20, 30],
    'min_samples_split' : [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4],
    'ccp_alpha' : [0.001, 0.002, 0.003, 0.004]
}

param_grid_rf = {
    'n_estimators' : [50, 100, 200],
    'max_depth' : [None, 10, 20, 30],
    'min_samples_split' : [2, 5, 10],
    'min_samples_leaf' : [1, 2, 4]
}

dt_grid_search = GridSearchCV(dt, param_grid_dt, cv=5, scoring='accuracy')
rf_grid_search = GridSearchCV(rf, param_grid_rf, cv=5, scoring='accuracy')


dt_grid_search.fit(X_train, y_train)
rf_grid_search.fit(X_train, y_train)

best_dt = dt_grid_search.best_estimator_
best_rf = rf_grid_search.best_estimator_

y_pred_dt = best_dt.predict(X_test)
y_pred_rf = best_rf.predict(X_test)

acc_dt = accuracy_score(y_test, y_pred_dt)
acc_rf = accuracy_score(y_test, y_pred_rf)

print(f"Test set accuracy Decision Tree: {acc_dt*100:.2f}%")
print(f"Test set accuracy Random Forest: {acc_rf*100:.2f}%")

from sklearn.model_selection import cross_val_score
scores_dt = cross_val_score(best_dt, X_train, y_train, cv=5)
print(f'Mean CV accuracy Decision Tree: {scores_dt.mean()}')
scores_rf = cross_val_score(best_rf, X_train, y_train, cv=5)
print(f'Mean CV accuracy Random Forest: {scores_rf.mean()}')
```

```python
# Import Library
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
```

```python

dt = pd.read_csv('dataset/diabetes.csv')
```

```python
data_filled = dt.fillna(0.0)
```

```python
X = data_filled.drop('Outcome', axis=1)
y = data_filled['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
regression = LogisticRegression(random_state=42)
svm = SVC(kernel='poly', degree=3, probability=True, random_state=42)
decision_tree = DecisionTreeClassifier(random_state=42)
```

```python
voting_classifier = VotingClassifier(
    estimators=[
        ('logistic_regression', regression),
        ('svc', svm),
        ('dt', decision_tree)
    ],
    voting='hard'
)
```

```python
voting_classifier.fit(X_train_scaled, y_train)
```

```python
y_pred = voting_classifier.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
print(f"Voting Ensemble accuracy (without hyperparameter tuning): {accuracy*100:.2f}%")
```

```python
param_grid_regression = {
    # 'penalty' : ['l1', 'l2', 'elasticnet', 'none'],
    'C' : [0.001, 0.01, 1, 10, 100],
    'solver' : ['liblinear', 'saga', 'lbfgs'],
    'max_iter' : [50, 100, 150, 200, 300],
    # 'l1_ratio' : [0.5, 0.7, 0.9]
}

param_grid_svm = {
    'C' : [0.01, 0.1, 1, 10, 100, 1000],
    'degree' : [2, 3, 4, 5],
    'gamma' : ['scale', 'auto'],
    'coef0' : [0, 0.5, 1],
}

param_grid_decision_tree = {
    'max_depth' : [None, 10, 20, 30],
    'min_samples_split' : [2, 5, 10]
}
```

```python
# List untuk menyederhanakan pemanggilan fit
log_reg = GridSearchCV(regression, param_grid_regression, cv=5, scoring='accuracy')
svc = GridSearchCV(svm, param_grid_svm, cv=5, scoring='accuracy')
dt = GridSearchCV(decision_tree, param_grid_decision_tree, cv=5, scoring='accuracy')

log_reg.fit(X_train_scaled, y_train)
svc.fit(X_train_scaled, y_train)
dt.fit(X_train_scaled, y_train)

best_log_reg = log_reg.best_estimator_
best_svc = svc.best_estimator_
best_dt = dt.best_estimator_
```

```python
voting_classifier = VotingClassifier(
    estimators=[
        ('regression', best_log_reg),
        ('svm', best_svc),
        ('decision_tree', best_dt)
    ],
    voting='soft'
)

voting_classifier.fit(X_train_scaled, y_train)
```

```python
y_pred = voting_classifier.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
print(f"Voting Ensemble accuracy (without hyperparameter tuning): {accuracy*100:.2f}%")

print(f'Best Logistic Regresion Model: {best_log_reg}')
print(f'Best SVM Model: {best_svc}')
print(f'Best Desicion Tree Model: {best_dt}')

from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)

print("Classification Report for Voting Classifier:\n", report)
```
